```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['class']=le.fit_transform(ds['class'])
```

```python
from scipy.stats import zscore
z=np.abs(zscore(ds))
```

```python
ds_n=ds[(z<3).all(axis=1)]
```

```python
ds_n.shape
```

```python
ds.shape
```

```python
ds=ds_n
ds.shape
```

```python
x=ds.drop(['class'],axis=1)
x.head()
```

```python
y=ds['class']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=63)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
KNR=KNeighborsClassifier()
DTR=DecisionTreeClassifier(random_state=63)
ABR=AdaBoostClassifier(random_state=63)
RFR=RandomForestClassifier(random_state=63)
GBR=GradientBoostingClassifier(random_state=63)
SV=SVC(random_state=63)
```

```python
models=[]
models.append(('DecisionTreeClassifier',DTR))
models.append(('KNeighborsClassifier',KNR))
models.append(('AdaBoostClassifier',ABR))
models.append(('RandomForestClassifier',RFR))
models.append(('GradientBoostingClassifier',GBR))
models.append(('SVC', SV))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
Model=[]
score = []
cvs=[]
MSE=[]
for name,model in models:
    print('*********************************',name,'*********************************')
    print('\n',model)
    Model.append(name)
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print('Accuracy_score = ',AS)
    score.append(AS*100)
    print('\n')
    sc = cross_val_score(model, x, y, cv=10, scoring='accuracy').mean()
    print('Cross_Val_Score = ',sc)
    cvs.append(sc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print('\n')
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
```

```python
result = pd.DataFrame({'Model': Model, 'Accuracy_score': score ,'Cross_val_score':cvs})
```
